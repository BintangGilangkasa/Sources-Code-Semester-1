```python
#Letak menginput Celcius ke Fahrenheit)
celcius = int(input('Berapa suhu Celcius = '))

#Rumus untuk mencari Suhu Celcius
fahrenheit = (celcius * 9/5) + 32

#input data
print (fahrenheit, "Fahrenheit" )
```

```python
#Letak menginput Fahrenheit ke Celcius
fahrenheit = int(input('Berapa suhu Fahreheit = '))

#Rumus untuk mencari Suhu Celcius
celcius = (fahrenheit - 32) * 5/9

#input data
print (celcius, "Celcius")
```
